```python
# shifted griewank function 
def griewank_function(x_values, data=griewank, dim=dimension, bias=f_bias):    
    F1 = 0
    F2 = 1
    for i in range(0,dim) :
        z  = x_values[i] - data[i]
        F1 = (z**2)/4000
        F2 = math.cos(z/((i+1)**.5))
        F1 += F1
        F2 *= F2
    return F1 - F2 + 1 + bias
```

```python
# keeping solutions stored
def store_solutions(iter):    
    global solutions_list
    solutions_list.append(griewank_function(iter))
```

```python
## results

start = timer()
x0 = np.random.uniform(x_min, x_max, dimension)
global solutions_list
solutions_list=[]
solution = minimize(griewank_function, x0, method='nelder-mead',options={'disp': True}, callback=store_solutions)

end = timer()
elapsed_time = end - start
print('Computation time (seconds): ', elapsed_time)
print(solution.x)
```
